```python
tmp229 = []
tmp228 = new_train['Target']
# tmp228
for i in tmp228:
    if i == 'CompPart':
        tmp229.append(1)
    elif i == 'Sub':
        tmp229.append(1)
    else:
        tmp229.append(0)
print(len(tmp229))
```

```python
# X = new_train[['month', 'year', 'CompPart', 'Comment', 'Sub', 'Disc', 'timestamp','Total_Num_User_Months','Zindi_Joining_Timestamp', 'comp_timestamp', 'Months_Since_Last_Comp',
#        'Months_Since_Joining_Zindi', 'Current_Active_Competitions',
#        'user_interests', 'User_ID_nunique', 'Total_Num_User_Months_mean',
#        'Total_Num_User_Months_max', 'Total_Num_User_Months_std',
#        'unique_user_count', 'user_interests_rank']]
X = new_train[['month', 'year', 'Comment', 'Disc', 'timestamp','Total_Num_User_Months','Zindi_Joining_Timestamp', 'comp_timestamp', 'Months_Since_Last_Comp',
       'Months_Since_Joining_Zindi', 'Current_Active_Competitions',
       'user_interests', 'User_ID_nunique', 'Total_Num_User_Months_mean',
       'Total_Num_User_Months_max', 'Total_Num_User_Months_std',
       'unique_user_count', 'user_interests_rank']]
y = new_train[['bin_target']]

X.fillna(0, inplace=True)
y.fillna(0, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=314, shuffle=True, stratify=y)

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

y_pred_log_reg = log_reg.predict(X_test)

f1_log_reg = f1_score(y_test, y_pred_log_reg, average='macro')
print(f1_log_reg)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

y_pred_knn = knn.predict(X_test)

f1_knn = f1_score(y_test, y_pred_knn, average='macro')
print(f1_knn)
```

```python
rf = RandomForestClassifier(n_estimators=5, random_state=314)
rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

f1_rf = f1_score(y_test, y_pred_rf, average='macro')
print(f1_rf)
```

```python
import xgboost as xgb
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split

dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

params = {
    'objective': 'binary:logistic',
    'eval_metric': 'logloss',
    'max_depth': 3,
    'eta': 0.1,
    'seed': 314
}

bst = xgb.train(params, dtrain, num_boost_round=100)

y_pred_prob = bst.predict(dtest)
y_pred = [1 if prob > 0.5 else 0 for prob in y_pred_prob]

f1_xgboost = f1_score(y_test, y_pred)
print(f1_xgboost)
```
